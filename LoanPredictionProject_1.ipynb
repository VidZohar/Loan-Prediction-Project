
# **LOAN PREDICTION - TRAIN DATASET**

```python
#IMPORTING LIBRARIES & DATASETS
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

train_dataset = pd.read_csv('train_dataset.csv')  #importing datasets (pandas)
```

```python
#FILLING MISSING DATA
train_values = {'Gender':'Male','Married':'Yes','Dependents':'0','Education':'Graduate','Self_Employed':'No','Property_Area':'Urban'}
train_dataset = train_dataset.fillna(value=train_values)
```

```python
x = train_dataset.iloc[:,1:-1].values
y = train_dataset.iloc[:, -1].values
```

```python
#CALCULATING MEAN FOR MISSING DATA
#(for 'ApplicantIncome', 'CoapplicantIncome', 
#'LoanAmount', 'Loan_Amount_Term', 'Credit_History')
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(x[:,5:10])
x[:,5:10] = imputer.transform(x[:,5:10])
```

```python
#ENCODING THE INDEPENDENT VARIABLE 
#(za stolpec 'Property_Area')
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer([('encoder', OneHotEncoder(),[10])], remainder = 'passthrough')
x = np.array(ct.fit_transform(x))
#zakodira 'Property_Area' v 001, 010, 100 
#in da te cifre na začetne tri stolpce
```

```python
#ENCODING THE INDEPENDENT VARIABLE 
#(za stolpec 'Dependents')
ct = ColumnTransformer([('encoder', OneHotEncoder(),[5])], remainder = 'passthrough')
x = np.array(ct.fit_transform(x))
#print(x_train[2])
```

```python
#ENCODING THE INDEPENDENT VARIABLE 
#(za ostale kategorijske stolpce)
x[x=='Male'] = 1
x[x=='Female'] = 0
x[x=='Yes'] = 1
x[x=='No'] = 0
x[x=='Graduate'] = 1
x[x=='Not Graduate'] = 0
```

```python
#ENCODING THE DEPENDENT VARIABLE
#from sklearn.preprocessing import LabelEncoder
#le = LabelEncoder()
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y) #'No' zakodira v 0 in 'Yes' zakodira v 1
#print(len(y)) #length = 614
#print(y)
```

```python
#SPLITING THE DATASETS INTO THE TRAIN SET AND TEST SET
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
#FEATURE SCALING
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)  
x_test = sc.transform(x_test)
```

```python
#TRAINING THE MODEL (LOGISTIC REGRESSION)
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)  #creating the model
classifier.fit(x_train, y_train)
```

```python
#PREDICTING THE TEST SET RESULTS
y_pred = classifier.predict(x_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))   #vidimo kaki je pravi rezultat in kaj vrne naš model
```

```python
#THE CONFUSION MATRIX
#confusion matrix shows the number of correct and incorrect predictions for 2 cases - 0 and 1
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
#accuracy_score = number of correct predictions / total number of observations in test set
accuracy_score(y_test, y_pred)  #pove (v procentih %) koliko pravilnih predikcij je naredil naš model
```
